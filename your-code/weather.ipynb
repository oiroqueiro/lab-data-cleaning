## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
```

```python
weather = pd.read_csv('..//weather-raw.csv')
weather
```

```python
weather_new = weather.melt(id_vars=['id','year','month','element'], var_name='day')
weather_new['day'] = weather_new['day'].str.replace("d","").astype(int)
weather_new
```

```python
weather_new = weather_new.pivot_table(index=['id','year','month','day'],columns='element',values='value')

weather_new
```

```python
weather_new.to_csv('weather.csv')
```
